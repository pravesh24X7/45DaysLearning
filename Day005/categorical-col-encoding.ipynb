### Encoding categorical data

```python
# import necessary libs

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/rashida048/Datasets/master/titanic_data.csv")
df.info()
```

In the above dataset, we can see categorical columns like `Sex`, `Embarked`, `Pclass`
Things to note that, columns like `Pclass` are already in interger format, therefore no need for Encoding such columns

For encoding `Sex` and `Embarked` different techniques are available:
+ OneHotEncoding
+ LableEncoding
+ BinaryEncoding

### Slight preprocessing ...

```python
remove_cols = [ 'Cabin', 'PassengerId', 'Ticket', 'Name' ]
df.drop(columns=remove_cols, inplace=True)
```

```python
df.Age = df.Age.fillna(df.Age.mean)
```

```python
df.Embarked = df.Embarked.fillna(df.Embarked.value_counts().sort_values(ascending=False).reset_index().iloc[0]['Embarked'])
```

### Performing Encoding on categorical columns

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False, drop='first')
encoded_cols = ohe.fit_transform(df[["Sex", "Embarked"]])
```

```python
encoded_cols
```

```python
ohe.get_feature_names_out(["Sex", "Embarked"])
```

We can perform the exact same task using pd.get_dummies() function.
But that's not recommended because sklearn.preprocessing.OneHotEncoder and requires manual coding while creating pipelines.
<br/>
Below is the demonstration of same.

```python
pd.get_dummies(df[["Sex", "Embarked"]]).head(5)
```

### Label Encoding example

```python
# creating a sample dataframe
test_df = pd.DataFrame({
    "age" : [10, 12, 13, 14, 15, 17],
    "marks": [ 50, 40, 36, 89, 23 , 90 ],
    "result": ["pass", "pass", "pass", "pass", "fail", "pass"]
})

test_df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
new_result = le.fit_transform(test_df.result)
```

```python
test_df["result"] = new_result
```

```python
test_df
```

### How is Binary encoding performed?
By applying custom logic, we can not just apply Binary, but other encoding as well.
<br/>
Below given code segment show the same.

```python
test_df["binary_result"] = test_df.marks.apply(
    (lambda mark: "Grade_A" if mark > 75 else "Grade_B")
)
```
