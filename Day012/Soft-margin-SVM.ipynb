### Day012 : Soft margin SVM [Cont.]

```python
# import necessary libs.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs
from sklearn.svm import SVC
from mlxtend.plotting import plot_decision_regions
from sklearn.inspection import DecisionBoundaryDisplay
```

### Generating synthetic dataset

```python
X, y = make_blobs(n_samples=100,
                 centers=2,
                 random_state=0,
                 cluster_std=0.90,)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap="winter")
```

From the above figure we can clearly see that the **Hard Margin SVM** fails, because of it's strict constraints.
<br/>
whereas, **Soft Margin SVM** will dominated here.

```python
# apply soft margin SVM

soft_svc_clf = SVC(C=2, kernel="linear")
soft_svc_clf.fit(X, y)
```

```python
plot_decision_regions(X, y, clf=soft_svc_clf, legend=2)
```

```python
DecisionBoundaryDisplay.from_estimator(
    soft_svc_clf, X,
    levels=[-1, 0, 1],
    plot_method="contour",
    alpha=0.5
)

plt.scatter(soft_svc_clf.support_vectors_[:, 0], soft_svc_clf.support_vectors_[:, 1], s=100, linewidth=4, edgecolors="k", facecolors="none")
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="winter", s=50)
plt.show()
```

Above figure shows, how **Soft margin SVC** handles the decision boundary by allowing some points to mis classify.
<br/>
This is where the *hyperparameter* `C` came into picture.
1. If we increase the value of `C` the `margin` decreases which leads to overfitting of model.
2. If we decrease the value of `C` the `margin` increases which leads to underfitting of model.
<br/>
<br/>
Therefore, it is necessary to find out the optimal values of `C` for which the model gives the best accuracy.
