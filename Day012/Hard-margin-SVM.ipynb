### SVM classifier

```python
# import necessary lib's

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.svm import SVC
from sklearn.datasets import make_blobs
from mlxtend.plotting import plot_decision_regions
from sklearn.inspection import DecisionBoundaryDisplay
```

### Generating synthetic dataset

```python
X, y = make_blobs(n_samples=100,
                  centers=2,
                  random_state=0,
                  cluster_std=0.60)

X.shape, y.shape
```

### How dataset looks like?

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="winter", s=50)
plt.show()
```

From the above figure, we can clearly see that, the **Hard Margin SVM** can easily classify this dataset.

```python
# applying hard margin SVM

svm_clf = SVC(kernel="linear", C=1)
svm_clf.fit(X, y)
```

### How Decision boundary looks like?

```python
plot_decision_regions(X, y, clf=svm_clf, legend=2)
```

Like we discussed above, simple **Hard Margin SVM** can easily classify the data points, let's plot our support vectors.

```python
DecisionBoundaryDisplay.from_estimator(
    svm_clf, X,
    plot_method="contour",
    levels=[-1, 0, 1],  # -1, 1 for support vectors and 0 for decision boundary.
    alpha=0.5,
)

plt.scatter(svm_clf.support_vectors_[:, 0], svm_clf.support_vectors_[:, 1], s=100, linewidth=1, facecolors="none", edgecolors="k")
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="winter", s=50)
plt.show()
```

Now, that we get our visualization for support vectors, we can conclude that the **Hard Margin SVM** classifier will perform terrible on the datasets where **non linear** decision boundary exists.
<br/>
This SVM classifier is not robust to outliers.
