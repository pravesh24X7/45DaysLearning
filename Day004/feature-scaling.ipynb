### Feature Scaling

```python
# import necessary libs.

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../titanic_clean.csv")
df.info()
```

### For demonstrating the feature scaling
first we'll select the Numerical columns.

```python
num_cols = df[['Age', 'Fare', 'FamilySize']]
num_cols.shape
```

### How original data looks like

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(data=num_cols)
plt.title("Before Feature Scaling")
plt.show()
```

### Applying Standardization

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
std_scaler = StandardScaler()
new_X = std_scaler.fit_transform(num_cols)
```

```python
sns.boxplot(data=new_X)
plt.title("After Feature Scaling")
plt.show()
```

### Interpreting the same results mathematically

```python
print("Before Scaling")
print(num_cols.describe())
```

```python
std_scaled_X = pd.DataFrame( new_X, columns=["Age", "Fare", "FamilySize"] )
print(std_scaled_X.describe())
```

### Applying Normalization

```python
min_max_scaler = MinMaxScaler()
X = min_max_scaler.fit_transform(num_cols)
```

```python
sns.boxplot(data=X)
plt.title("After Feature Scaling")
plt.show()
```

### Interpreting the results of Normalization mathematically,

Here, we can clearly see that only the scale is changed, According to that `mean` and `std` values are not 0 and 1.

```python
norm_scaled_X = pd.DataFrame( X, columns=["Age", "Fare", "FamilySize"] )
print(norm_scaled_X.describe())
```
