### KNN Algorithm

```python
# import necessary lib's 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score
from mlxtend.plotting import plot_decision_regions
```

### Loading dataset

```python
X, y = load_iris(return_X_y=True)

figure = plt.figure(figsize=(12, 5))
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="winter", s=50)
plt.subplot(122)
plt.scatter(X[:, 2], X[:, 3], c=y, cmap="viridis", s=50)
plt.show()
```

Observation from the above plots, we can easily classify if we train model on `feature3` and `feature4`.
<br/>
But here, we'are training model using `KNN` algorithm, so we'll use all 4 features to train our model, Although we can train model on the subset of features, but in this case we're not doing that.

### Split dataset for training and testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    random_state=42,
                                                    test_size=0.2)
```

### Model training

```python
knn_clf = KNeighborsClassifier(n_neighbors=5)
knn_clf.fit(X_train, y_train)
```

```python
y_pred = knn_clf.predict(X_test)
f1_score(y_test, y_pred, average="weighted"), accuracy_score(y_test, y_pred)
```

Our `KNN` classifier easily classify the test datapoints, which we can clearly see from the `f1_score` and `accuracy_score`.
