### Regularization (Ridge and Lasso)

```python
# import necessary libs.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import Ridge
```

```python
from sklearn.linear_model import Lasso, LinearRegression, ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import fetch_california_housing
```

```python
california_housing = fetch_california_housing(as_frame=True)
df = california_housing.data
y = california_housing.target
df.shape, y.shape
```

First we'll perform all the necessary preprocessing steps which we did before while learning Linear regression.
<br/>
These steps includes:
1. Removing `AvgBedrms` columns.
2. Performing `log` transformation on `MedInc`, `Population` and `target` columns.
3. Scaling operation on rest of the dataset.

```python
print(df.columns)
```

```python
df.drop(columns=['AveBedrms'], inplace=True)
df['MedInc'] = np.log1p(df["MedInc"])
df["Population"] = np.log1p(df["Population"])
y = np.log1p(y)
```

```python
std_scaler = StandardScaler()
X = std_scaler.fit_transform(df)
```

### Splitting dataset for training and testing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    random_state=42,
                                                    test_size=0.3)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### Simple Linear Regression
Frist we'll fit the normal regression model and check it's `RMSE` value.

```python
simple_linear_regression = LinearRegression(n_jobs=-1)
simple_linear_regression.fit(X_train, y_train)
```

### Evaluating Simple linear regression performance

```python
y_prediction_simple_linear_regression = simple_linear_regression.predict(X_test)

y_prediction_simple_linear_regression = np.expm1(y_prediction_simple_linear_regression)
y_test_actual = np.expm1(y_test)
print("RMSE of Simple Linear Regression : ", mean_squared_error( y_test_actual, y_prediction_simple_linear_regression )**0.5)
```

### Ridge Regression (L2 Regularization)
Trying Ridge regression and check it's RMSE score.

```python
l2_regularization = Ridge(alpha=5)
l2_regularization.fit(X_train, y_train)
```

### Evaluating L2 regularization performance

```python
y_predictions_l2_regularization = l2_regularization.predict(X_test)

y_predictions_l2_regularization = np.expm1(y_predictions_l2_regularization)
print("RMSE of L2 Regularization : ", mean_squared_error(y_test_actual, y_predictions_l2_regularization) ** 0.5)
```

### Lasso Regression (L1 Regularization)
Trying Lasso and validating it's performance.

```python
l1_regularization = Lasso(alpha=0.25)
l1_regularization.fit(X_train, y_train)
```

### Evaluating L1 Regularization performance

```python
y_predictions_l1_regularization = l1_regularization.predict(X_test)
y_predictions_l1_regularization = np.expm1(y_predictions_l1_regularization)

print("RMSE of L1 Regularization : ", mean_squared_error(y_test_actual, y_predictions_l1_regularization) ** 0.5)
```

### ElasticNet Regression
This is hybird regularizer that utilizes both the `Ridge` and `Lasso`

```python
elastic_net_regularizer = ElasticNet(alpha=5e-6, l1_ratio=0.000005)
elastic_net_regularizer.fit(X_train, y_train)
```

### Evaluting ElasticNet Regularizer

```python
y_predictions_elastic_net = elastic_net_regularizer.predict(X_test)
y_predictions_elastic_net = np.expm1(y_predictions_elastic_net)

print("RMSE of Elastic Net : ", mean_squared_error(y_test_actual, y_predictions_elastic_net) ** 0.5)
```

**Final Regularization Order**:
<br/>
ElasticNet < Ridge < Simple Linear Regression < Lasso
